```
import nltk
nltk.download('stopwords')
```

```
#Remove the stop word and unwanted text 
import re
from nltk.corpus import stopwords
def clean_text(text,remove_stopwords=True):

  if True:
    #formatting words and remove unwanted characters
    text=re.sub(r'https?:\/\/.*[\r\n]*', '', str(text),  
                  flags=re.MULTILINE)
    text=re.sub(r'\<a href', ' ', str(text))
    text=re.sub(r'&amp;', '', str(text))
    text=re.sub(r'[_"\-;%()|+&=*%.,!?:#$@\[\]/]', ' ', str(text))
    text=re.sub(r'<br />', ' ', str(text))
    text=re.sub(r'\'', ' ', str(text))

    #optinally remove stopwords
    if remove_stopwords:
      text=text.split()
      stops=set(stopwords.words('english'))
      text=[w for w in text if not w in stops]
      text=" ".join(text)
    return text

new_text=clean_text(public_tweets)

```
